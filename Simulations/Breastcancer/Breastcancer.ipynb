```python
import numpy as np
import pandas as pd
from time import time
import matplotlib.pyplot as plt
from matplotlib.pyplot import cm
from sklearn import manifold, datasets
import seaborn as sns
from isumap import isumap
```

```python
n_neighbors=30
data=pd.read_csv('./../Dataset_fils/data.csv')
data = data.drop('id',axis=1)
data=data.drop('Unnamed: 32',axis=1)
data['diagnosis'] = data['diagnosis'].map({'M':1,'B':0})
X = data.values
colors=data['diagnosis']
```

```python
Model1,cluster= isumap(data= X, k=30, d=2, normalize= True, distBeyondNN = True, verbose=True, switchToBigDataAt=1500, dataIsDistMatrix=False,
                dataIsGeodesicDistMatrix = False, MDSmethod="metric", saveDistMatrix = False)
```

```python
fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111)
fig.add_axes(ax)
ax.scatter(
    Model1[:, 0], Model1[:, 1], s=10 ,c=colors , cmap=plt.cm.rainbow
)
```
